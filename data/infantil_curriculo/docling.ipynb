```python
import os
import pandas as pd
import re

folder = "."
arquivos = [f for f in os.listdir(folder) if f.endswith('.xlsx') and not f.startswith('infantil_curriculo_normalizado')]

linhas = []

def encontrar_coluna(df, padrao):
    for col in df.columns:
        nome = re.sub(r"\s+", " ", col.upper().strip())
        if re.search(padrao, nome):
            return col
    return None

for arquivo in arquivos:
    caminho = os.path.join(folder, arquivo)
    df_raw = pd.read_excel(caminho, header=None)
    # Extrai o título da linha 2 (índice 1)
    titulo = ""
    for cell in df_raw.iloc[1]:
        if isinstance(cell, str) and "-" in cell:
            titulo = cell
            break
    match = re.search(r"-\s*([^-\n]+?)\s*-\s*([^-\n]+)", titulo)
    ano = match.group(1).strip() if match else ""
    eixo = match.group(2).strip().upper() if match else ""

    # Lê a planilha pulando as 3 primeiras linhas (header na linha 3)
    df = pd.read_excel(caminho, header=2)
    df.columns = [col.strip() for col in df.columns]
    df = df.fillna(method="ffill")  # Preenche células mescladas

    col_objetivo = encontrar_coluna(df, r"OBJETIVO.*APRENDIZAGEM")
    col_exemplos = encontrar_coluna(df, r"EXPERI(Ê|E)NCIA|EXEMPLO")

    for _, row in df.iterrows():
        objetivo = str(row[col_objetivo]).replace('\n', ' ').replace('\r', ' ') if col_objetivo else ""
        exemplos = str(row[col_exemplos]).replace('\n', ' ').replace('\r', ' ') if col_exemplos else ""

        linhas.append({
            "SEGMENTO": "EDUCAÇÃO INFANTIL",
            "ANO": ano,
            "EIXO": eixo,
            "OBJETIVO DE APRENDIZAGEM": objetivo,
            "EXEMPLOS": exemplos
        })

df_final = pd.DataFrame(linhas)
df_final.to_excel("infantil_curriculo_normalizado.xlsx", index=False)
print("Arquivo infantil_curriculo_normalizado.xlsx criado com sucesso!")
```
